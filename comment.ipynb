```python
#remove punctuation
import re
import emoji
def remove_punc(text):
    whitespace = re.compile(u"[\s\u0020\u00a0\u1680\u180e\u202f\u205f\u3000\u2000-\u200a]+", re.UNICODE)
    bangla_fullstop = u"\u0964"
    punctSeq   = u"['\"“”‘’]+|[.?!,…]+|[:;]+"
    punc = u"[৷–()—,$%^&*+={}\[\]:\"|\'\~`<>/,¦!?½£¶¼©⅐⅑⅒⅓⅔⅕⅖⅗⅘⅙⅚⅛⅜⅝⅞⅟↉¤¿º;-]+"
    text= whitespace.sub(" ",text).strip()
    text= re.sub(punctSeq, " ", text)
    text = re.sub(bangla_fullstop, " ",text)
    text = re.sub(punc, " ", text)
    text = re.sub('\w*\d\w*', " ", text)
    text = re.sub(emoji.get_emoji_regexp(), "", text)
    return text
```

```python
cdf['Clean']= cdf['Comment'].apply(lambda x: remove_punc(x))
cdf
```

```python
#Tokenize
cdf['tokenized_sents'] = cdf.apply(lambda row: nltk.word_tokenize(row['Clean']), axis=1)
cdf
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state = 10)
```

# DT

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(max_depth=50)#criterion='entropy',max_depth=50
dtree.fit(x_train,y_train)
y_pred=dtree.predict(x_test)
y_pred
```

# MLP

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(hidden_layer_sizes=50, activation='relu',max_iter=100)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
y_pred
```
